```python
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from scipy import stats

diabetes=datasets.load_diabetes()
x=diabetes.data
y=diabetes.target

x2=sm.add_constant(x)
est=sm.OLS(y,x2)
est2=est.fit()
print(est2.summary())
```

```python
df=pd.read_csv('HR_Comma_Sep.csv')
df.shape
## drop department columns
#create dummy variables for salary
subdf=df.drop(['sales','last_evaluation','number_project'], axis=1)
subdf
subdf2=pd.get_dummies(subdf, prefix='salary', drop_first=True)
subdf2
```

```python
from sklearn.model_selection import train_test_split
x=subdf2.drop('left', axis=1)

y=subdf2.left
train_x, test_x, train_y, test_y=train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
train_x2=sm.add_constant(train_x)
ols=sm.OLS(train_y,train_x2)
ols2=ols.fit()
print(ols2.summary())
```

```python
train_x2=sm.add_constant(train_x)
logit=sm.Logit(train_y,train_x2)
logit2=logit.fit()
print(logit2.summary())
```
